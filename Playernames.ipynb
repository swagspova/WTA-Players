```python
import pandas as pd
from string import ascii_uppercase as alphabet
import pickle
```

```python
USO_23 = pd.read_html('https://en.wikipedia.org/wiki/2023_US_Open_%E2%80%93_Women%27s_singles#Draw')
```

```python
dict_table = {}
for i in range(4,12,1):
    df = USO_23[i]
    df.rename(columns= {2:'Player', 1: 'Misc'}, inplace=True)
    df = df[['Misc','Player']]
    df = df.drop(df.index[:2])
    dict_table[f'Section {i-3}'] = df

    print(dict_table)
```

```python
dict_table['Section 8']
```

```python
with open('dict_table', 'wb') as output:
    pickle.dump(dict_table,output)
```

```python
# Combine all DataFrames in the dictionary into one
combined_df = pd.concat(dict_table.values(), keys=dict_table.keys())

# Reset the index of the combined DataFrame
combined_df.reset_index(level=0, inplace=True)
combined_df.rename(columns={'level_0': 'Section'}, inplace=True)

# Display the combined DataFrame
print(combined_df)
```
